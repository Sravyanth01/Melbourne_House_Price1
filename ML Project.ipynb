```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import ShuffleSplit
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 20, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model1 = LinearRegression()
model1.fit(X_train, Y_train)
model2 = neighbors.KNeighborsRegressor(n_neighbors = 200)
model2.fit(X_train, Y_train)
model3 = DecisionTreeRegressor(random_state = 0) 
model3.fit(X_train, Y_train)
model4 = RandomForestRegressor(n_estimators = 100, random_state = 0)
model4.fit(X_train, Y_train)
```

```python
Y_predict1 = model1.predict(X_test)
Y_predict2 = model2.predict(X_test)
Y_predict3 = model3.predict(X_test)
Y_predict4 = model4.predict(X_test)
```

```python
mse1 = mean_squared_error(Y_predict1,Y_test)
mse2 = mean_squared_error(Y_predict2,Y_test)
mse3 = mean_squared_error(Y_predict3,Y_test)
mse4 = mean_squared_error(Y_predict4,Y_test)

print("MSE1: ", mse1 )
print("MSE2: ", mse2 )
print("MSE3: ", mse3 )
print("MSE4: ", mse4 )
```

```python
R2Score1  = r2_score(Y_predict1,Y_test)
R2Score2 = r2_score(Y_predict2,Y_test)
R2Score3  = r2_score(Y_predict3,Y_test)
R2Score4  = r2_score(Y_predict4,Y_test)

print("R2Score1: ", R2Score1 )
print("R2Score2: ", R2Score2 )
print("R2Score3: ", R2Score3 )
print("R2Score4: ", R2Score4 )
```

```python
cv = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
```

```python
#linear Regression
scores1 = cross_val_score(model1, X_train, Y_train,cv=cv)
print('Accuracy: %.3f (%.3f)' % (mean(scores1), std(scores1)))
```

```python
#K Nearest Neighbour
scores2 = cross_val_score(model2, X_train, Y_train,cv=cv, n_jobs=-1)
print('Accuracy: %.3f (%.3f)' % (mean(scores2), std(scores2)))
```

```python
#Descsion Tree
scores3 = cross_val_score(model3, X_train, Y_train,cv=cv, n_jobs=-1)
print('Accuracy: %.3f (%.3f)' % (mean(scores3), std(scores3)))
```

```python
#Random Forest
scores4 = cross_val_score(model4, X_train, Y_train,cv=cv, n_jobs=-1)
print('Accuracy: %.3f (%.3f)' % (mean(scores4), std(scores4)))
```

```python
scores = [scores1,scores2,scores3,scores4]
names = ["LR","KNR","DTR","RFR"]
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(scores)
ax.set_xticklabels(names)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor


def find_best_model_using_gridsearchcv(X,Y):
    algos = {
        'R_F_R': {
            'model': RandomForestRegressor(),
            'params': {
        
                'criterion': ['squared_error', 'friedman_mse',],
                'n_estimators':[5000,10000],
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,Y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,Y)
```
